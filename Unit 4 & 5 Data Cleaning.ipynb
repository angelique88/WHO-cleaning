# Explore missing data

```python
# Find out number of missing values in each column

missing = raw.isna().sum()
missing = missing.to_frame().reset_index()
missing = missing.rename(columns= {'index': 'col', 0: 'missing vals'})
missing
```

```python
missing.describe()
```

Drop columns with over 7000 missing values out of 9008.

```python
drop_col = missing[missing["missing vals"] >7000]
drop_col
```

```python
#List of columns with too many missing values

to_drop = list(drop_col["col"])
to_drop
```

```python
#Dropping the columns with too many missing values

raw2 = raw.drop(to_drop, axis = 1)
```

```python
raw2.head(20)
```

# Check dataypes for all columns

```python
raw2.dtypes
```

- Most float64 can convert to int64. Except for 'Wealth index score'
- 'Date of interview women (CMC)', 'Date of birth of woman (CMC)', 'Date of marriage (CMC)' values do not really make sense so we will drop them.
- 'Children ever born' is duplicated 3 times.

```python
# Dropping columns that do not add value

raw2 = raw2.drop(['Date of interview women (CMC)', 'Date of birth of woman (CMC)', 'Date of marriage (CMC)'], axis = 1)
```

```python
float_col = raw2.select_dtypes('float64').columns.to_list()
float_col = float_col[:-1]
float_col
```

```python
raw2[float_col].head()
```

```python
# Remove na values and replace with median

for x in float_col:
    raw2[x] = raw2[x].fillna(raw2[x].median())
```

```python
raw2[float_col].isna().sum()
```

```python
# Drop duplicate columns

raw2 = raw2.drop(["Children ever born"], axis = 1)
```

```python
raw2["Children ever born"] = raw2["Children surviving"] + raw2["Children dead"]
raw2["Children ever born"] = raw2["Children ever born"].astype(int)
```

```python
# Convert columns from float64 to int64 
for x in float_col:
    raw2[x] = raw2[x].astype(int)
```
